```python
import math
import torch
import torch.nn as nn
import torch.nn.functional as F


class SimplePTCMIL(nn.Module):
    def __init__(self, d_model=512, n_prompts=5, n_classes=2, nhead=8):
        super().__init__()
        self.d_model = d_model
        self.n_prompts = n_prompts

        # learnable tokens
        self.prompt = nn.Parameter(torch.randn(1, n_prompts, d_model) * 0.02)  # [1, K, D]
        self.cls = nn.Parameter(torch.randn(1, 1, d_model) * 0.02)             # [1, 1, D]

        # Step 1: global Transformer (官方实现)
        g_layer = nn.TransformerEncoderLayer(
            d_model=d_model, nhead=nhead, dim_feedforward=4 * d_model,
            dropout=0.0, batch_first=True, norm_first=True
        )
        self.global_encoder = nn.TransformerEncoder(g_layer, num_layers=1)

        # Step 3: local Transformer (参数共享，官方实现)
        l_layer = nn.TransformerEncoderLayer(
            d_model=d_model, nhead=nhead, dim_feedforward=4 * d_model,
            dropout=0.0, batch_first=True, norm_first=True
        )
        self.local_encoder = nn.TransformerEncoder(l_layer, num_layers=1)

        # Step 4: merge 权重
        self.merge_score = nn.Linear(d_model, 1)

        # Step 5: slide classifier
        self.classifier = nn.Linear(d_model, n_classes)

    def forward(self, patch_feats):
        """
        patch_feats: [1, N, D]
        """
        B, N, D = patch_feats.shape
        assert B == 1, "这里做单张 slide 模拟"
        assert D == self.d_model

        # Step 1: [cls, prompt, patches] -> global transformer
        seq = torch.cat([self.cls, self.prompt, patch_feats], dim=1)  # [1, 1+K+N, D]
        out = self.global_encoder(seq)

        cls1 = out[:, :1, :]                        # [1,1,D]
        p1 = out[:, 1:1 + self.n_prompts, :]       # [1,K,D]
        e1 = out[:, 1 + self.n_prompts:, :]        # [1,N,D]

        # Step 2: patch-to-prompt soft clustering
        sim = torch.matmul(e1, p1.transpose(-1, -2)) / math.sqrt(D)   # [1,N,K]
        A = F.softmax(sim, dim=-1)                                     # [1,N,K]
        hard_idx = A.argmax(dim=-1)[0]                                 # [N]

        # Step 3+4: per-cluster local Transformer + weighted merge -> prototype
        prototypes = []
        cluster_sizes = []
        for c in range(self.n_prompts):
            mask = (hard_idx == c)
            h1c = e1[:, mask, :]                                       # [1, Mc, D]
            cluster_sizes.append(int(mask.sum().item()))

            if h1c.shape[1] == 0:
                hpc = p1[:, c, :]                                      # [1,D]
            else:
                h2c = self.local_encoder(h1c)                          # [1, Mc, D]
                r = self.merge_score(h2c).squeeze(-1)                  # [1, Mc]
                w = F.softmax(r, dim=-1).unsqueeze(-1)                 # [1, Mc, 1]
                hpc = (w * h2c).sum(dim=1)                             # [1,D]
            prototypes.append(hpc)

        H_P = torch.stack(prototypes, dim=1)                           # [1,K,D]

        # Step 5: cls + prototypes pooling
        H_final = torch.cat([cls1, H_P], dim=1)                        # [1,1+K,D]
        slide_repr = H_final.mean(dim=1)                               # [1,D]
        logits = self.classifier(slide_repr)                           # [1,C]

        aux = {
            "A": A,
            "hard_idx": hard_idx,
            "cluster_sizes": cluster_sizes,
            "cls1": cls1,
            "H_P": H_P,
            "H_final": H_final,
        }
        return logits, aux


def main():
    torch.manual_seed(42)
    device = "cpu"

    # 可改为 12000；CPU 下可能较慢/占内存
    N_PATCHES = 12000
    D = 512
    K = 5
    N_CLASSES = 2

    # Step 0: 模拟 patch 特征提取
    patch_feats = torch.randn(1, N_PATCHES, D, device=device)  # [1,N,512]
    y_true = torch.tensor([0], device=device)                  # 假设真值 LUAD=0

    model = SimplePTCMIL(d_model=D, n_prompts=K, n_classes=N_CLASSES).to(device)
    optimizer = torch.optim.Adam(model.parameters(), lr=1e-4)

    # 前向（更新前）
    logits_before, aux_before = model(patch_feats)
    prob_before = F.softmax(logits_before, dim=-1).detach().cpu().numpy()[0]
    loss_before = F.cross_entropy(logits_before, y_true)

    print("=== 模拟结果（更新前） ===")
    print("patch_feats:", tuple(patch_feats.shape))
    print("A shape:", tuple(aux_before["A"].shape))
    print("cluster sizes:", aux_before["cluster_sizes"])
    print("H_P shape:", tuple(aux_before["H_P"].shape))
    print("H_final shape:", tuple(aux_before["H_final"].shape))
    print("pred prob [class0, class1]:", prob_before)
    print("loss_before:", float(loss_before.item()))

    # Step 6: 反向更新（task-aware）
    optimizer.zero_grad()
    loss_before.backward()
    optimizer.step()

    # 前向（更新后）
    logits_after, aux_after = model(patch_feats)
    prob_after = F.softmax(logits_after, dim=-1).detach().cpu().numpy()[0]
    loss_after = F.cross_entropy(logits_after, y_true)

    print("\n=== 1步更新后 ===")
    print("cluster sizes:", aux_after["cluster_sizes"])
    print("pred prob [class0, class1]:", prob_after)
    print("loss_after:", float(loss_after.item()))


if __name__ == "__main__":
    main()
```
